```python
import numpy as np
import matplotlib.pyplot as plt
from parse_file import parse_Gaussian_file
from plot_functions import plot_spectra_nm
```

```python
#function to convert machine absorption to ext.coeff, I assumed cavity length=1
def abs(abs):
    return [(i/0.000002) for i in abs]
```

```python
#Spectra Rhodamine 6G 
nm_rho6G_NAOH, abs_rho6G_NAOH = np.loadtxt('experimental/200220_abs_0,1MNaOH_2,0uM_Rho6G.TXT', unpack=True)
nm_rho6G_EtOH, abs_rho6G_EtOH = np.loadtxt('experimental/200220_abs_NaOH-EtOH_2,0uM_Rho6G.TXT', unpack=True)
nm_rho6G_MECN, abs_rho6G_MECN = np.loadtxt('experimental/200220_abs_NaOH-MeCN_2,0uM_Rho6G.TXT', unpack=True)
energR6G, oscR6G = parse_Gaussian_file('theoretical/R6GES.log')
energR6G_wat, oscR6G_wat = parse_Gaussian_file('theoretical/R6G_CPCM_waterES.log')
energR6G_eth, oscR6G_eth = parse_Gaussian_file('theoretical/R6G_CPCM_ethanolES.log')
energR6G_acet, oscR6G_acet = parse_Gaussian_file('theoretical/R6G_CPCM_acetoES.log')
```

```python
plt.plot(nm_rho6G_NAOH, abs(abs_rho6G_NAOH), label='NaOH')
plt.plot(nm_rho6G_EtOH, abs(abs_rho6G_EtOH), label='EtOH')
plt.plot(nm_rho6G_MECN, abs(abs_rho6G_MECN), label='MECN')
plot_spectra_nm(energR6G, oscR6G, legend='Gas-phase')
plot_spectra_nm(energR6G_wat, oscR6G_wat, legend='PCM-water')
plot_spectra_nm(energR6G_eth, oscR6G_eth, legend='PCM-ethanol')
plot_spectra_nm(energR6G_acet, oscR6G_acet, legend='PCM-acetonitrile')
plt.legend()
plt.ylabel('$\\frac{L}{molcm}$')
plt.title('Rhodamine 6G in High pH (0.1 M NaOH)')
plt.xlim(300, 650)
plt.ylim(0,200000)
plt.show()
```

```python
plt.close()
```
