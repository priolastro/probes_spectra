```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#function to convert machine absorption to ext.coeff, I assumed cavity length=1
def abs(abs):
    return [(i/0.000002) for i in abs]
```

```python
nm_rho6G_NAOH, abs_rho6G_NAOH = np.loadtxt('200220_abs_0,1MNaOH_2,0uM_Rho6G.TXT', unpack=True)
nm_rho6G_EtOH, abs_rho6G_EtOH = np.loadtxt('200220_abs_NaOH-EtOH_2,0uM_Rho6G.TXT', unpack=True)
nm_rho6G_MECN, abs_rhoMECN = np.loadtxt('200220_abs_NaOH-MeCN_2,0uM_Rho6G.TXT', unpack=True)
```

```python
plt.plot(nm_rho6G_NAOH, abs(abs_rho6G_NAOH), label='NaOH')
plt.plot(nm_rho6G_EtOH, abs(abs_rho6G_EtOH), label='EtOH')
plt.plot(nm_rho6G_MECN, abs(abs_rhoMECN), label='MECN')
plt.legend()
plt.ylabel('$\\frac{L}{molcm}$')
plt.title('Rhodamine 6G in High pH (0.1 M NaOH)')
plt.xlim(350, 600)
plt.ylim(0,90000)
plt.show()
```
